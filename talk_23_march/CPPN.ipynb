```python
from keras.models import Sequential
from keras.layers import Dense
from keras.initializers import RandomUniform

import numpy as np
import matplotlib.pyplot as plt
```

```python
init = RandomUniform(minval=-3, maxval=3)

model = Sequential()
model.add(Dense(20, activation='tanh', input_shape=(256, 256, 3), kernel_initializer=init, bias_initializer=init))

model.add(Dense(20, activation='tanh', kernel_initializer=init, bias_initializer=init))
model.add(Dense(20, activation='tanh', kernel_initializer=init, bias_initializer=init))
model.add(Dense(20, activation='tanh', kernel_initializer=init, bias_initializer=init))

model.add(Dense(3, activation='sigmoid'))
```

```python
model.summary()
```

```python
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)
```

```python
xy = np.meshgrid(x, y)
```

```python
tensor_data = []

for i, xi in enumerate(np.sin(x)):
    for j, yj in enumerate(np.cos(y)):
        tensor_data.append([xi, yj, xi**2 + yj**2])
        
tensor_data = np.asarray(tensor_data)
```

```python
tensor_data.shape
```

```python
output = model.predict(tensor_data)
```

```python
reshaped = np.reshape(output, (1000, 1000, 3))
```

```python
plt.figure(figsize=(15, 15))
plt.axis('off')
plt.imshow(reshaped)
```
