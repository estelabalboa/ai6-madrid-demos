## Grammatical evolution

```python
import numpy as np
```

```python
# f: R5 -> R
```

```python
# Definimos la función que queremos aproximar, para poder comparar luego con la función que encuentra

def f(x):
    return np.cos(x[4]) * x[1]- x[3] + x[0] - x[2]
```

```python
inputs = [list(10 * np.random.random(5) - 5) for i in range(10000)]
```

```python
outputs = list(map(lambda u: f(u), inputs))
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(data=inputs)
```

```python
data.columns = ['x0', 'x1', 'x2', 'x3', 'x4']
```

```python
data['response'] = outputs
```

```python
data[:8000].to_csv('datasets/Train.txt', sep='\t', index=None)
data[8000:].to_csv('datasets/Test.txt', sep='\t', index=None)
```

```python
#respuesta del PonyGE
def modelo(x):
    return x[0]-x[3]-x[2]-np.sin(x[4]-np.tanh(x[1]))-np.sin(x[4]-np.sin(np.tanh(x[1]))-np.tanh(x[1]))-np.sin(x[1]-np.tanh(np.sin(x[4]-np.tanh(x[1])))-np.sin(x[4]-np.tanh(x[1])-np.sin(np.sin(x[4]-x[4]-x[1])+np.tanh(np.tanh(np.tanh(np.tanh(x[1]))))))-np.tanh(x[1]))
```

```python
y_pred = list(map(lambda u: modelo(u), inputs))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 10))
plt.xlabel('Y predecida')
plt.ylabel('Y real')
plt.scatter(y_pred, outputs, color='blue', s=0.25)
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
fixed_dims = [ 9.40737545e-01,  1.69036095e-01,  4.65696739e+00, -2.34077746e+00, -1.67828359e+00]
```

```python
def generate_test_data(fixed, x_axis, free_dim):
    test_data = []
    for i in x_axis:
        fixed[free_dim] = i
        test_data.append(np.asarray(fixed))    
    return test_data
```

```python
test_data = generate_test_data(fixed_dims, x, 4)
```

```python
for i in range(5):
    test_data = generate_test_data(fixed_dims, x, i)
    plt.plot(x, list(map(lambda u: modelo(u), test_data)), label='modelo')
    plt.plot(x, list(map(lambda u: f(u), test_data)), label='real')
    plt.legend()
    plt.show()
```
