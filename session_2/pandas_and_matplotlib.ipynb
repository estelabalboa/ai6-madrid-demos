```python
n = 50

instance = df.iloc[n]
labels = list(instance.index)
values = list(instance.values)

labels.append('TARGET')
values.append(target[n])
```

```python
def get_values(position):

    instance = df.iloc[position]
    l = list(instance.index)
    v = list(instance.values)

    l.append('TARGET')
    v.append(target[position])
    
    return l, v
```

```python
plt.figure(figsize=(20, 12))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.xticks(rotation=45)
    lab, val = get_values(i+20)
    plt.bar(lab, val, color=colors)
```

# One hot encoding

```python
one_hot_river = pd.get_dummies(df['CHAS'])
one_hot_river.columns = ['CHAS_{}'.format(i) for i in range(one_hot_river.shape[1])]
```

```python
df['CHAS'] = one_hot_river
```

```python
#aÃ±adimos esas dos columnas a nuestro dataframe
df = df.join(one_hot_river)
#quitamos la columna original
#df.drop('CHAS', axis=1, inplace=True)
df = df.drop('CHAS', axis=1)
```

```python
x = df.values
y = target.values
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x)
x = scaler.transform(x)
```

```python
max(y)
```

```python
y = y/50
```

# Train/Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

#Â Machine learning

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg = RandomForestRegressor()
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
R2 = r2_score(y_pred, y_test)
```

```python
R2
```

```python
plt.plot([0, 1], [0, 1])
plt.scatter(y_test, y_pred)
plt.title('R2-score: {}'.format(R2))
```

```python
reg.score(X_test, y_test)
```
